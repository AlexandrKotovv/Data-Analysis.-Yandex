А там где статус помещения соответстсвует аппартаментам продавцы целенаправленно активировали значение True. Поэтому заполним пропуски значением False.

```python
# check
# пропущенные значения, бары

def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
            .plot(kind = 'barh', figsize = (19,6), rot = 0, legend = False, fontsize = 16)
            .set_title('Пример' + "\n", fontsize = 22, color = 'SteelBlue')    
        );    
    except:
        print('пропусков не осталось :) ')
```

```python
pass_value_barh(data)
```

```python
# для дальнейшего исследования проведём категоризацию общей площади помещений

def categorize_area(area):
    try:
        if  area <= 25:
            return 'J'
        elif 25 < area <= 35:
            return 'I'
        elif 35 < area <= 45:
            return 'H'
        elif 45 < area <= 55:
            return 'G'
        elif 55 < area <= 65:
            return 'F'
        elif 65 < area <= 75:
            return 'E'
        elif 75 < area <= 85:
            return 'D'
        elif 85 < area <= 95:
            return 'C'
        elif 95 < area <= 105:
            return 'B'
        elif area > 105:
            return 'A'
    except:
        pass
data['total_area_category'] = data['total_area'].apply(categorize_area)
data.head(9)
```

```python
# Для квартир-студий площадь кухни назначим 0

data.loc[data['studio'] == True, 'kitchen_area'] = 0
```

Достоверного способа заполнить пропуски, не зная год постройки дома или его серию, практически невозможно.  
Поэтому используем следующую методику:

```python
# по каждой категории общей площади считаем, какая медианная доля общей площади - жилая, 
# и заполним пропуски таким же значением

for i in data['total_area_category'].unique():
    ratio = (data[data['total_area_category'] == i]['living_area'] / data[data['total_area_category'] == i]['total_area']).median()
    data['living_area'] = data['living_area'].fillna(value = data['total_area'] * ratio)
```

```python
data['living_area'].isna().sum() # проверим не осталось ли пропусков
```

```python
# проверим, что не вышло жилой площади больше общей

data[data['living_area'] > data['total_area']]['total_area'].count()
```

```python
# проверим, что получилось

data['living_area'].describe()
```

```python
# по каждой комнатности и высотности дома считаем, какая медианная доля общей площади - кухня,  
# и заполняем пропуски таким же значением

for i in data['rooms'].unique():
    for k in data['floors_total'].unique():
        try:
            ratio = (data[data['rooms'] == i]['kitchen_area'] / data[data['rooms'] == i]['total_area']).median()
            data['kitchen_area'] = data['kitchen_area'].fillna(value = data['total_area'] * ratio)
        except:
             pass
```

```python
data['kitchen_area'].isna().sum() # проверим остались ли пропуски
```

```python
# проверим, что получилось

data['kitchen_area'].describe()
```

```python
# посмотрим сколько стало объектов, где сумма площади кухни и жилой составляют более 90% общей площади объекта,
# и скорректируем жилую площадь. Было 475, стало 580

(data['total_area']*0.9 <= data['living_area'] + data['kitchen_area']).sum()
```

Для таких объектов примем такое эмпирическое правило - жилая площадь равна 80% общей за вычетом кухни.

```python
area_check = data.total_area*0.9 - data.kitchen_area - data.living_area 
data.loc[area_check < 0, 'living_area'] = (data.total_area - data.kitchen_area) * 0.8
```

```python
# посмотрим сколько осталось таких объектов после корректировки

(data['total_area']*0.9 < data['living_area'] + data['kitchen_area']).sum()
```

Добавим столбец, в котором будет указан тип этажа помещения (значения — «первый», «последний», «другой»)

```python
def categorize_floor(row):
    try:
        if  row['floor'] == 1:
            return 'первый'
        elif row['floor'] == row['floors_total']:
            return 'последний'
        elif 1 < row['floor'] < row['floors_total']:
            return 'другой'
    except:
        pass
data['floor_category'] = data.apply(categorize_floor, axis=1)
print(data.head())
```

### Изучим какие факторы влияют на цену

```python
# рассмотрим влияние общей площади квартиры на её цену
# построим диаграмму рассеивания
total_area_pivot = data.pivot_table(index='total_area', values='last_price').reset_index()
total_area_pivot.plot(x='total_area', y='last_price', kind='scatter',figsize=(10, 8),alpha=0.3);
plt.title('Диаграмма рассеяния — Общая площадь — Стоимость квартиры');
```

На графике видим, что в дипазоне квартир с общей площадью до 100 кв.м.(таких подавляющее большинство) существует положительная линейная зависимость.

```python
# посчитаем коэффициент корреляции
total_area_pivot['total_area'].corr(total_area_pivot['last_price'])
```

Таким образом, график показывает и положительный коэффициент корреляции=0.87 это подтверждает, что определённо есть положительная зависимость между общей площадью квартиры и её ценой - чем больше площадь, тем больше цена.

```python
# рассмотрим влияние жилой площади квартиры на её цену
# построим диаграмму рассеивания
living_area_pivot = data.pivot_table(index='living_area', values='last_price').reset_index()
living_area_pivot.plot(x='living_area', y='last_price', kind='scatter',figsize=(10, 8),alpha=0.3);
plt.title('Диаграмма рассеяния — Жилая площадь — стоимость квартиры');
```

```python
# посчитаем коэффициент корреляции
living_area_pivot['living_area'].corr(living_area_pivot['last_price'])
```

Как и в случае с общей площадью квартиры между жилой площадью и ценой квартиры есть положительная линейная зависимость, но уже чуть слабее.

### Посчитаем среднюю цену одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений.  
   Выделим населённые пункты с самой высокой и низкой стоимостью квадратного метра.

```python
top_10 = data['locality_name'].value_counts().head(10)
# top_10
top_10 = ['Санкт-Петербург','поселок Мурино','поселок Шушары',
          'Всеволожск','Пушкин','Колпино','поселок Парголово',
          'Гатчина','деревня Кудрово','Выборг']

city_av_price = data.query('locality_name  in @top_10')\
                .pivot_table(index = 'locality_name',\
                             values = 'price_per_metr')
city_av_price = city_av_price.sort_values(by='price_per_metr',ascending=False)
city_av_price.style.format("{:,.1f}")    
```

Среди десяти населённых пунктов с максимальным числом объявлений самая высокая стоимость кв.м. в Санкт-Петербурге,  
самая низкая стоимость кв.м. в Выборге.
